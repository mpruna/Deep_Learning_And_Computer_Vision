### Open CV commands description:

    * cv2.imshow() ----> Use the function cv2.imshow() to display an image in a window. The window automatically fits to the image size.

    * cv2.waitKey() ---->  is a keyboard binding function. Its argument is the time in milliseconds. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. If 0 is passed, it waits indefinitely for a key stroke

    * cv2.destroyAllWindows() ----> simply destroys all the windows we created. If you want to destroy any specific window, use the function cv2.destroyWindow() where you pass the exact window name as the argument.
    
[OpenCV cmds reference](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_image_display/py_image_display.html)

### Convert to gray color

```python
gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Watching TV in a CAR!', gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
gray_image.shape
```

```python
### Convert RGB to HSV

image=cv2.imread('watchingTV_moving_car.jpg')
cv2.imshow('Watching TV in the car!', image)

hsv_image=cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV image!', hsv_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
![HSV IMAGE]()
```

### Extract the HSV layers from image

```python
plt.imshow(hsv_image[:, :, 0])
plt.title('Hue channel')
```

```python
plt.imshow(hsv_image[:, :, 1])
plt.title('Saturation channel')
```

```python
plt.imshow(hsv_image[:, :, 2])
plt.title('Value channel')
```

### Split and merge R, G, B channels

```python
image=cv2.imread('watchingTV_moving_car.jpg')
B,G,R=cv2.split(image)
```

```python
B.shape
```

```python
G.shape
```

```python
R.shape
```

```python
cv2.imshow('R channel!', R)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow('R channel!', G)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Unexpected results

![Gray Image]()

This is because OpenCV interprets one layer as default gray, it does not understand if it's `RED` or `Green`

```python
### Create our own image with Blue channel

To do this we need to specify that the other two channels are `zero`
```

```python
zeros=np.zeros(image.shape[:2],dtype='uint8')
cv2.imshow("Blue channel!", cv2.merge([B, zeros, zeros]))
cv2.waitKey()
cv2.destroyAllWindows()
```

### Merge back the image

```python
image_merged=cv2.merge([B,G,R])
cv2.imshow('Merged_image',image_merged)
cv2.waitKey()
cv2.destroyAllWindows()
```
