Let's try to simulate a real world scenario where we want to approximate the vehicle speed based on road bumps. In real world situations we use LIDAR to get this kind of data

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
vehicleSpeed=np.random.normal(40.0,10.0, 1000)                  ### Create Sample data for vehicle Speed
bumpiness=(1 - (vehicleSpeed/70.0)) +np.random.normal(0,0.1,1000)  ### Create Bump data sample
```

**vehicleSpeed**:
* create 1000 samples, distributed around 40 value with a standard deviation of 10

**bumpiness**:
* to create some sort of a link between these sets we must create this set fom vehicleSpeed. This approch will create a set good for linear regression

```python
sns.regplot(vehicleSpeed,bumpiness, fit_reg=False) # By default, a linear regression fit is drawn, you can remove it with fit_reg=False
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err=stats.linregress(vehicleSpeed, bumpiness)
```

### Create the linear regression equation 

```python
def predict(x):
    return slope *x + intercept          '''y =  +  x  math formula for linear regression'''

sns.set()

fitline=predict(vehicleSpeed)
plt.scatter(vehicleSpeed, bumpiness)     '''create scatter plot'''
plt.plot(vehicleSpeed, fitline, c='r')   '''plot the linear regersion line''' 
plt.show()
```

```python
sns.regplot(vehicleSpeed,bumpiness)
```

We calculated the error performance by determining the **RSS(RESIDUAL SUM OF SQUARED)** or simply put if y1' the actual value and y(x1) is the approximation the then RSS is the also known as the sum of squared residuals (SSR) or the sum of squared errors of prediction
