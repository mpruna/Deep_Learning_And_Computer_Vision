```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image_color=cv2.imread('images/watchingTV_moving_car.jpg')
image_red_1=cv2.imread('images/red_car_2.jpg')
image_red_2=cv2.imread('images/red_night.jpg')
```

```python
def cv2_plot(image):
    cv2.imshow('Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
### Plto all images colored;red 1; red 2
```

```python
cv2_plot(image_color)
plt.imshow(image_color)
```

```python
cv2_plot(image_red_1)
plt.imshow(image_red_1)
```

It's somewhat expected to have a blue color while ploting with matplotlib because `CV2` uses BGR instead on RBG color scale.

```python
cv2_plot(image_red_2)
plt.imshow(image_red_2)
```

```python
def get_histogram(image):
    colors=('b','g','r')
    
    for i, col in enumerate(colors):
        histogram2=cv2.calcHist([image],[i],None,[255],[0,255])
        plt.plot(histogram2, color=col)
        plt.xlim([0,255])
        
    plt.show()
```

```python
get_histogram(image_color)
```

It's expected to have multiple colors and to have blue as the primary color

```python
get_histogram(image_red_1)
```

```python
get_histogram(image_red_2)
```

It's somewhat expected to have the highest value pixels around 0=black as the background of the picture is black
