### Convert to gray scale

```python
image_gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
image_gray.shape
```

### GET CANNY EDGE DETECTION

```python
threshold_1=20
threshold_2=250

image_canny=cv2.Canny(image_gray, threshold_1,threshold_2)

cv2.imshow('Canny Edge detecion', image_canny)
cv2.waitKey()
cv2.destroyAllWindows()
```

### FAST FEATURE DETECTION

```python
fast=cv2.FastFeatureDetector_create()
keypoints=fast.detect(image_gray,None)

print('Number of points detected', len(keypoints))

image=cv2.drawKeypoints(image, keypoints, None, color=(255,0,0))

cv2.imshow('Fast Feature detection', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### ORB FEATURE DETECTION

```python
orb=cv2.ORB_create()
keypoints=orb.detect(image_gray,None)

keypoints, descriptors=orb.compute(image_gray,keypoints)
print('Number of keypoints created', len(keypoints))

image=cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2.imshow('Orb feature show', image)
cv2.waitKey()
cv2.destroyAllWindows()
```
