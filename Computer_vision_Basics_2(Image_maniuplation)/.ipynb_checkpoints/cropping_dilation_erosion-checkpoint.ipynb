```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

### Load image

```python
image=cv2.imread('Future_SelfDriving.jpg')
```

```python
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
```

```python
print('Height:',str(height)+'\n','Width:', str(width))
```

### Perform cropping

```python
# Top left corners of the desired area

ho=int(height*0.3)
w0=int(width*0.3)

# Bottom right of the desired area

h1=int(height*1)
w1=int(width*1)

image_cropped=image[ho:h1,w0:w1]
```

```python
cv2.imshow('Cropped Image', image_cropped)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# If you want to see the actual distance translated
# Don't worry about the color, recall openCV uses BGR instead of RGB
plt.imshow(image_cropped)
```

```python
plt.imshow(image)
```

# PERFORM DILATION AND EROSION OPERATIONS

```python
image=cv2.imread("Speed_Sign_View_2.jpg")
```

```python
kernel=np.ones((8,8),np.uint8)

image_erode=cv2.erode(image, kernel, iterations=2)
image_dilate=cv2.dilate(image, kernel, iterations=2)

cv2.imshow('Ogirinal Image', image)
cv2.imshow('Dilate', image_erode)
cv2.imshow('Erode',image_dilate)

cv2.waitKey()
cv2.destroyAllWindows()
```
