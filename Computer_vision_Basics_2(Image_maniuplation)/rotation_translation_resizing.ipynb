```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
image=cv2.imread('Future_SelfDriving.jpg')
```

### Load Display Image

```python
image=cv2.imread('Future_SelfDriving.jpg')
cv2.imshow('Original image', image)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height , width = image.shape[:2]
```

```python
print('Height:', str(height) +'\n' + 'Width:' , str(width))
```

### Perform Rotations

```python
M_rotation=cv2.getRotationMatrix2D((width/2, height/2), 30, 0.6)  #Provide the coords(x,y), angle, scale i.e scale down 0.6
rotated_image=cv2.warpAffine(image, M_rotation,(width, height)) #rotate image based on rotated matrix 

cv2.imshow('Rotated', rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Perform Translations

```python
translational_matrix=np.float32([[1,0,100],
                                [0,1,50]])

translated_image=cv2.warpAffine(image,translational_matrix,(width,height))
#translated_mage=cv2.imshow('Translated image', translated_image)
#cv2.waitKey()
#cv2.destroyAllWindows()
```

```python
plt.imshow(translated_image)
```

### Perform Resizing

```python
resize_image = cv2.resize(image, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_CUBIC)
cv2.imshow('Resized Image', resize_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
